```python
import itk
import sys
import matplotlib.pyplot as plt
import numpy as np
import SimpleITK as sitk
```

```python
inputFileName = "test_imagen.jpeg"
outputFileName = "test_imagen2.jpeg"
```

# Lectura imagenes

```python
ImageType = itk.Image[itk.UC, 2]

# Leer la imagen con el lector
reader = itk.ImageFileReader[ImageType].New()
reader.SetFileName(inputFileName)
reader.Update()

# Obtener la imagen del lector
image = reader.GetOutput()

# Obtener la información de la imagen
size = image.GetLargestPossibleRegion().GetSize()
spacing = image.GetSpacing()
origin = image.GetOrigin()

# Imprimir la información de la imagen
print("Size:", size)
print("Spacing:", spacing)
print("Origin:", origin)
```

```python
temp=10
median = itk.median_image_filter(image, radius=temp)
outputFileName = f"test_imagen{temp}.jpeg"
itk.imwrite(median, outputFileName)
```

```python
# Convertir la imagen ITK a una matriz numpy
np_image = itk.GetArrayFromImage(image)

# Mostrar la imagen usando matplotlib
plt.imshow(np_image)#, cmap='gray')
plt.show()
```

```python
# Convertir la imagen ITK a una matriz numpy
np_image = itk.GetArrayFromImage(median)

# Mostrar la imagen usando matplotlib
plt.imshow(np_image, cmap='gray')
plt.show()
```
