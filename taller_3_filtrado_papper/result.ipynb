```python
import itk

# Read the MNC image
mnc_image = itk.imread('data/t1_ai_msles2_1mm_pn0_rf20.mnc')

# Get the image data as a numpy array
data = itk.array_from_image(mnc_image)

# Print the shape of the data array
print(data.shape)

# Plot a slice of the image
import matplotlib.pyplot as plt
plt.imshow(data[:,:,30], cmap='gray')
plt.show()
```

```python
import nibabel as nib
import itk

# Read the MNC file using nibabel
mnc_img = nib.load('data/t1_icbm_normal_1mm_pn9_rf20.mnc')

# Get the image data as a numpy array
data = mnc_img.get_fdata()

# Convert the numpy array to an ITK object
itk_img = itk.image_view_from_array(data)

# Print the image information
print(itk_img)

# Plot a slice of the image
import matplotlib.pyplot as plt
plt.imshow(itk.array_view_from_image(itk_img)[:,:,30], cmap='gray')
plt.show()
```
