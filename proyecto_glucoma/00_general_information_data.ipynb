```python
import os
from PIL import Image
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import random
```

```python
folder_path = 'data'
```

```python
# Count number of files in folder
num_files = len([f for f in os.listdir(folder_path) if os.path.isfile(os.path.join(folder_path, f))])
print("Number of files in folder:", num_files)
```

```python
# Get pixel size of images in folder
pixel_sizes = []
for filename in os.listdir(folder_path):
    if os.path.isfile(os.path.join(folder_path, filename)):
        try:
            with Image.open(os.path.join(folder_path, filename)) as img:
                pixel_sizes.append(img.size)
        except:
            pass

```

```python
df_pixel = pd.DataFrame(pixel_sizes, columns=['pixel1', 'pixel2'])
```

#### Plot distribution

```python
# Alternatively, you can use the plot method with kind='hist' parameter
df_pixel['pixel1'].plot(kind='hist')

# Set the title and labels for the plot
plt.title('pixel x distribution')
plt.xlabel('Values')
plt.ylabel('Frequency')

# Show the plot
plt.show()
```

```python
# Alternatively, you can use the plot method with kind='hist' parameter
df_pixel['pixel2'].plot(kind='hist')

# Set the title and labels for the plot
plt.title('pixel y distribution')
plt.xlabel('Values')
plt.ylabel('Frequency')

# Show the plot
plt.show()
```

### Plot random images

```python
def plot_image_stats(file_path):
    # Load the image
    img = Image.open(file_path)

    # Convert the image to a numpy array
    img_array = np.array(img)

    # Plot the image using matplotlib
    plt.imshow(img_array)
    plt.axis('off')

    # Show the plot
    plt.show()

    # Get the principal statistics of the image
    img_size = img.size
    img_mean = np.mean(img_array)
    img_std = np.std(img_array)

    # Print the principal statistics of the image
    print("Image size:", img_size)
    print("Image mean:", img_mean)
    print("Image standard deviation:", img_std)
```

```python
listImages = [f'{folder_path}/{f}' for f in os.listdir(folder_path) if os.path.isfile(os.path.join(folder_path, f))]
```

```python
random_element = random.choice(listImages)
plot_image_stats(file_path= random_element)
```
